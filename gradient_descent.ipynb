## Basic Gradient Descent practise

```python
# functions(x)) = (x+2)**3
# gradient becomes as = 3(x+2) ** 2
import matplotlib.pyplot as plt
def function(x):
    return (x+2)**2

def gradient(x):
    return 2*(x+2)

x = 0
epochs = 30
lr = 0.01

losses = []
weights = []

for _ in range(epochs):
    weights.append(x)
    ot = function(x)
    losses.append(ot)
    gr = gradient(x)
    x = x - lr * gr
    # print(x,gr)
print(weights,losses,)
plt.plot(weights,losses,marker='o')
plt.xlabel("W")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

## Gradient Descent for Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

#sample data points = 
np.random.seed(42)
x =  2 * np.random.rand(100, 1)
y = 3 * x + 4 
```

```python
y[0],x[0], (3*x[0]+4)
```

```python
np.random.randn(1)
```

```python
w = np.random.randn(1)
losses = []
b = 0
lr = 0.001

training_steps = 1000

for epoch in range(training_steps):
    y_pred = x.dot(w) + b
    error = y - y_pred
    dl_dw = -2 * np.mean(x * error)
    dl_db = -2 * np.mean(error)
    w -= lr*dl_dw
    b -= lr*dl_db

    loss = np.mean(error) ** 2
    losses.append(loss)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss={loss:.4f} | w={w[0]:.4f} | b={b:.4f}")

plt.plot(np.arange(training_steps),losses,)
plt.xlabel('training_steps')
plt.ylabel('Loss')
plt.show()
```
